```python

import matplotlib
import pandas as pd
import numpy as np
```

```python
Data=pd.read_csv(f'/MDcnty.csv')
print(Data.head())
```

```python
import matplotlib.pyplot as plt
Data=pd.read_csv(f'/MDcnty.csv')
x=Data['Life Expectancy']
y=Data['Census Tract Number']
dataframe=pd.DataFrame({'x':x,'y':y})
plt.scatter(dataframe.x,dataframe.y),plt.xlabel("Life Expectancy"),plt.ylabel("Census Tract"),plt.title("Geography vs Life Miami-Dade")
plt.ylim(0,210)

matplotlib.pyplot.savefig('MDcnty.jpeg')
matplotlib.pyplot.savefig('MDcnty.png')
```
