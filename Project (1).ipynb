```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('/content/MDcnty.csv')
x = Data['Life Expectancy']
y = Data['Census Tract Number']
plt.scatter(x, y, c=x, cmap='RdYlGn', edgecolors='k', alpha=0.7)
plt.xlabel("Life Expectancy")
plt.ylabel("Census Tract")
plt.title("Geography vs Life Miami-Dade")
plt.ylim(0, 210)
colorbar = plt.colorbar()
colorbar.set_label('Life Expectancy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
Data = pd.read_csv('/content/MD Income.csv')
x = Data['Median Income in Thousands']
y = Data['Census Tract']
plt.scatter(x, y, c=x, cmap='RdYlGn', edgecolors='k', alpha=0.7)
plt.xlabel("Income")
plt.ylabel("Census Tract")
plt.title("Geography vs Median Income Miami-Dade County")
plt.xticks(np.arange(4.8, 250, 40))
plt.ylim(0, 210)
colorbar = plt.colorbar()
colorbar.set_label('Median Income in Thousands')
plt.show()
```
